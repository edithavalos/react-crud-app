**Ejercicio 1**:

Cargar el archivo california_housing_train.csv y crear un gráfico de dispersión (scatter plot) de median_income vs. median_house_value.

```python
import pandas as pd
import matplotlib.pyplot as plt

california_housing_train = pd.read_csv('sample_data/california_housing_train.csv')

plt.scatter(california_housing_train['median_income'], california_housing_train['median_house_value'])
```

**Ejercicio 2**:

Usando el mismo conjunto de datos, crear un histograma de la columna median_house_value con 50 bins y agregarle títulos y etiquetas a los ejes.

```python
plt.hist(california_housing_train['median_house_value'], bins=50)
plt.title('Histograma de median_house_value')
plt.xlabel('median_house_value')
plt.ylabel('Frecuencia')
```

**Ejercicio 3**:

Generar un gráfico de barras que muestre el número de casas (households) en cada rango de edad (housing_median_age), agrupando las edades en bins de 5 años.

```python
bins = range(0, int(california_housing_train['housing_median_age'].max()) + 5, 5)
california_housing_train['age_range'] = pd.cut(california_housing_train['housing_median_age'], bins=bins)

households_by_age = california_housing_train.groupby('age_range')['households'].sum()

plt.bar(households_by_age.index.astype(str), households_by_age.values)
plt.xlabel('Rango de Edad')
plt.ylabel('Número de Casas')
plt.title('Número de Casas por Rango de Edad')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Ejercicio 4**:

Crear un mapa de calor (heatmap) que muestre la correlación entre las variables numéricas del DataFrame california_housing_train.csv.

```python
import seaborn as sns

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('sample_data/california_housing_train.csv')

corr_matrix = df.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Mapa de Calor de Correlación')
plt.show()
```

**Ejercicio 5**:

Usar el archivo mnist_train_small.csv para mostrar la imagen del dígito en la primera fila, convirtiendo los valores de los píxeles en una matriz de 28x28 y visualizándola con imshow.

```python
import pandas as pd
import matplotlib.pyplot as plt

mnist_data = pd.read_csv('sample_data/mnist_train_small.csv')

first_image_pixels = mnist_data.iloc[0, 1:]

image_matrix = first_image_pixels.values.reshape(28, 28)

plt.imshow(image_matrix, cmap='gray')
plt.title('Imagen del dígito en la primera fila')
plt.show()
```

**Ejercicio 6**:

Crear subplots que muestren imágenes de los dígitos del 0 al 9, seleccionando una muestra de cada uno del conjunto de datos mnist_train_small.csv.

```python
fig, axs = plt.subplots(2, 5, figsize=(15, 6))

for i in range(10):
  digit_row = mnist_data[mnist_data['6'] == i].iloc[0]

  pixel_values = digit_row[1:].values

  image_matrix = pixel_values.reshape(28, 28)

  row = i // 5
  col = i % 5
  axs[row, col].imshow(image_matrix, cmap='gray')
  axs[row, col].set_title(f'Dígito {i}')
  axs[row, col].axis('off')

plt.tight_layout()
plt.show()
```

**Ejercicio 7**:

Cargar los datos de anscombe.json y crear gráficos de dispersión para cada uno de los cuatro conjuntos de datos de Anscombe, incluyendo la línea de regresión en cada gráfico.

```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
from scipy import stats

try:
  # Attempt to open the file
  with open('anscombe.json', 'r') as f:
    data_list = json.load(f)

  df1 = pd.DataFrame.from_dict(data_list[0], orient='index').T
  df2 = pd.DataFrame.from_dict(data_list[1], orient='index').T
  df3 = pd.DataFrame.from_dict(data_list[2], orient='index').T
  df4 = pd.DataFrame.from_dict(data_list[3], orient='index').T

  # Convert 'X' and 'Y' columns to numeric
  df1['X'] = pd.to_numeric(df1['X'])
  df1['Y'] = pd.to_numeric(df1['Y'])
  df2['X'] = pd.to_numeric(df2['X'])
  df2['Y'] = pd.to_numeric(df2['Y'])
  df3['X'] = pd.to_numeric(df3['X'])
  df3['Y'] = pd.to_numeric(df3['Y'])
  df4['X'] = pd.to_numeric(df4['X'])
  df4['Y'] = pd.to_numeric(df4['Y'])


  # Function to generate plots
  def plot_anscombe(df, title):
    plt.figure(figsize=(8, 6))
    # Access the columns using the correct label 'X' and 'Y'
    plt.scatter(df['X'], df['Y'])
    slope, intercept, r_value, p_value, std_err = stats.linregress(df['X'], df['Y'])
    line = slope*df['X']+intercept
    plt.plot(df['X'], line, color='red')
    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()

  # Generate plots for each dataset
  plot_anscombe(df1, 'Conjunto de datos 1')
  plot_anscombe(df2, 'Conjunto de datos 2')
  plot_anscombe(df3, 'Conjunto de datos 3')
  plot_anscombe(df4, 'Conjunto de datos 4')

except FileNotFoundError:
  print("Error: File 'anscombe.json' not found. Please check the file path or download the file.")
except IndexError:
    print("Error: The index is out of range for the data list. Please check the data structure in 'anscombe.json'.")
```

**Ejercicio 8**:

Usando california_housing_train.csv, crear un boxplot de median_house_value para diferentes categorías de ocean_proximity (nota: si esta columna no existe, crear categorías basadas en la ubicación geográfica).

```python
if 'ocean_proximity' in california_housing_train.columns:
  plt.figure(figsize=(12, 6))
  plt.boxplot([california_housing_train[california_housing_train['ocean_proximity'] == category]['median_house_value']
             for category in california_housing_train['ocean_proximity'].unique()],
            labels=california_housing_train['ocean_proximity'].unique())
  plt.xlabel('Ocean Proximity')
  plt.ylabel('Median House Value')
  plt.title('Boxplot de Median House Value por Ocean Proximity')
  plt.xticks(rotation=45, ha='right')
  plt.tight_layout()
  plt.show()
else:
  bins = [32, 34, 36, 38, 40, 42]
  labels = ['Zona 1', 'Zona 2', 'Zona 3', 'Zona 4', 'Zona 5']
  california_housing_train['latitude_category'] = pd.cut(california_housing_train['latitude'], bins=bins, labels=labels)

  plt.figure(figsize=(12, 6))
  plt.boxplot([california_housing_train[california_housing_train['latitude_category'] == category]['median_house_value']
             for category in california_housing_train['latitude_category'].unique()],
            labels=california_housing_train['latitude_category'].unique())
  plt.xlabel('Categoría de Latitud')
  plt.ylabel('Median House Value')
  plt.title('Boxplot de Median House Value por Categoría de Latitud')
  plt.xticks(rotation=45, ha='right')
  plt.tight_layout()
  plt.show()
```

**Ejercicio 9**:

Crear una gráfica de líneas que muestre la tendencia de median_house_value en función de la latitude, ordenando los datos por latitude

```python
if 'latitude' in california_housing_train.columns:
  california_housing_train = california_housing_train.sort_values('latitude')

  plt.figure(figsize=(12, 6))
  plt.plot(california_housing_train['latitude'], california_housing_train['median_house_value'], marker='o')
  plt.xlabel('Latitude')
  plt.ylabel('Median House Value')
  plt.title('Tendencia de Median House Value en función de Latitude')
  plt.grid(True)
  plt.show()
else:
  print("La columna 'latitude' no existe en el DataFrame.")
```

**Ejercicio 10**:

Generar un gráfico de pastel (pie chart) que muestre la proporción de viviendas en diferentes rangos de median_income, por ejemplo: bajos ingresos (<2), ingresos medios (2-5), ingresos altos (>5).

```python
income_labels = ['Bajos ingresos (<2)', 'Ingresos medios (2-5)', 'Ingresos altos (>5)']
income_ranges = [(0, 2), (2, 5), (5, float('inf'))]

income_counts = []
for start, end in income_ranges:
  count = california_housing_train[(california_housing_train['median_income'] >= start) &
                                   (california_housing_train['median_income'] < end)].shape[0]
  income_counts.append(count)

plt.figure(figsize=(8, 8))
plt.pie(income_counts, labels=income_labels, autopct='%1.1f%%', startangle=90)
plt.title('Proporción de Viviendas por Rango de Ingresos')
plt.show()
```
